```python
import pandas as pd
import random

exercises_list = ['Squats', 'Deadlifts', 'Bench Press', 'Pull-Ups', 'Overhead Press', 
                  'Lunges', 'Rows', 'Planks', 'Leg Press', 'Lat Pulldowns', 
                  'Chest Flyes', 'Bicep Curls', 'Tricep Dips', 'Hammer Curls', 
                  'Russian Twists', 'Calf Raises', 'Face Pulls', 'Leg Curls', 
                  'Dumbbell Rows', 'Incline Press', 'Side Planks', 'Arnold Press', 
                  'Seated Rows', 'Leg Extensions', 'Push-Ups', 'Crunches']

def format_exercise(exercises, sets, reps):
    if reps == 3:
        rep_format = '10-8-6'
    elif reps == 4:
        rep_format = '12-10-8-6'
    else:
        rep_format = 'Unknown Rep Format'

    exercise_str = ', '.join(exercises)
    return f"{exercise_str}, {sets} sets ({rep_format} reps)"

data = {
    'Plan': [f'Plan {chr(65 + i)}' for i in range(26)],  # 'A' to 'Z'
    'Exercise': [format_exercise(random.sample(exercises_list, k=random.randint(3, 4)),
                                 sets, reps) for sets, reps in zip([3, 4, 3, 4, 3, 4, 3, 4, 3, 4, 3, 4, 3, 4, 3, 4, 3, 4, 3, 4, 3, 4, 3, 4, 3, 4],
                                                                 [3, 4, 3, 4, 3, 4, 3, 4, 3, 4, 3, 4, 3, 4, 3, 4, 3, 4, 3, 4, 3, 4, 3, 4, 3, 4])],
}

df = pd.DataFrame(data)
print(df)
```
